We are building a logistic regression model that will try to predict the next month returns based off the past month return, past 3 months return, past 6 months return and past 9 months return.

```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime as dt
import plotly.express as px
```

```python
tickers = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]['Symbol'].str.replace('.', '-')
start = dt.datetime(2013, 1, 1)
starting_date = dt.datetime(2014, 1, 1)
end = dt.datetime.now()
interval = '1mo'
```

```python
def get_n_mtl_returns(df, periods):
    for i in periods:
        df[f'{i}_mt_ret'] = df['Adj Close'].pct_change(i).shift()
    return df
```

```python
def prep_stock_df(stock_df):
    stock_df = get_n_mtl_returns(stock_df, [1, 3, 6, 9]).dropna()
    stock_df['returns'] = stock_df['Adj Close'].pct_change()
    stock_df = stock_df[stock_df.index >= starting_date]
    return stock_df
```

```python
df = pd.DataFrame()

for ticker in tickers:
    try:
        stock_df = yf.download(ticker, start=start, end=end, interval=interval).iloc[:, -2:-1]
        stock_df['ticker'] = ticker
        stock_df['date'] = stock_df.index
        stock_df = prep_stock_df(stock_df)

        if df.empty:
            df = stock_df
        else:
            df = pd.concat([df, stock_df])
        print(ticker, 'is done')
        

    except Exception as e:
        print(ticker, e)
```

```python
# Calculate quantiles
#quantiles = df['returns'].quantile([0.25, 0.5, 0.75])
#q1 = quantiles[0.25]
#median = quantiles[0.5]
#q3 = quantiles[0.75]


# Define bins
#bins = [-float('inf'), q1, median, q3, float('inf')]
#labels = ['Very Low', 'Low', 'High', 'Very High']
df['category'] = pd.qcut(df['returns'], q=6, duplicates='drop', labels=['-0.833, -0.058', '-0.058, -0.0171', '-0.0171, 0.0124', '0.0124, 0.0413', '0.0413, 0.0812', '0.0812, 4.898'])

df = df.dropna().reset_index()
```

```python
def analyze_returns():# Basic statistics
    mean_returns = df['returns'].mean()
    median_returns = df['returns'].median()
    std_returns = df['returns'].std()
    quantiles_returns = df['returns'].quantile([0.25, 0.5, 0.75])

    print("Mean Returns:", mean_returns)
    print("Median Returns:", median_returns)
    print("Standard Deviation of Returns:", std_returns)
    print("Quantiles of Returns:", quantiles_returns)

    # Histogram
    fig_hist = px.histogram(df, x='returns', title='Histogram of Returns', labels={'returns': 'Returns'})
    fig_hist.update_layout(xaxis_title='Returns', yaxis_title='Count')
    fig_hist.show()

    # Box plot
    fig_box = px.box(df, y='returns', title='Box Plot of Returns', labels={'returns': 'Returns'})
    fig_box.update_layout(yaxis_title='Returns')
    fig_box.show()

    # Descriptive statistics summary
    descriptive_stats = df['returns'].describe()
    print(descriptive_stats)
```

```python
X = df[['1_mt_ret', '3_mt_ret', '6_mt_ret', '9_mt_ret']]
y = df['category']
y = y.astype('category')
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
model = LogisticRegression(max_iter=1000)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
model = model.fit(X_train, y_train)
y_preds = pd.DataFrame(model.predict(X_test))
```

```python
dict(zip(X_train.columns, model.coef_[0])) #Prints an array of all regressor values (b1 and b2, or as many bs as your model has)
#model.intercept_
```

```python
y_preds = y_preds.rename(columns={0: 'predicted_category'})
y_preds.index = X_test.index
test_df = pd.concat([X_test, y_test, y_preds], axis=1)
merged_df = pd.merge(df[['ticker', 'date','returns']], test_df, left_index=True, right_index=True)
```
